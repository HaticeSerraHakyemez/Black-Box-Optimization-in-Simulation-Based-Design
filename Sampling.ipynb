## Random Sampling


```python
# Re-importing necessary libraries and defining the Eggholder function again due to code execution state reset.
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.stats import qmc

def eggholder_function(x, y):
    term1 = -(y + 47) * np.sin(np.sqrt(abs(x/2 + (y + 47))))
    term2 = -x * np.sin(np.sqrt(abs(x - (y + 47))))
    return term1 + term2

# Domain boundaries
x_min, x_max = -512, 512
y_min, y_max = -512, 512

# Generating a meshgrid for plotting
x_values = np.linspace(x_min, x_max, 1000)
y_values = np.linspace(y_min, y_max, 1000)
X, Y = np.meshgrid(x_values, y_values)
Z = eggholder_function(X, Y)

# Number of samples
num_samples = 50

# Random sampling
random_samples = np.random.rand(num_samples, 2)  # Generate random samples in 2D

# Scale the random samples to the domain
random_samples_scaled = qmc.scale(random_samples, (x_min, y_min), (x_max, y_max))

# Evaluate the Eggholder function at the random samples
random_function_values = eggholder_function(random_samples_scaled[:, 0], random_samples_scaled[:, 1])

# Plotting
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

# Surface plot of the Eggholder function
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.5)

# Scatter plot of the random samples
ax.scatter(random_samples_scaled[:, 0], random_samples_scaled[:, 1], random_function_values, color='red', s=50)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Eggholder Function')

plt.title("Random Sampling of the Eggholder Function")
plt.show()
```

## Monte Carlo Sampling

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the Eggholder function
def eggholder_function(x, y):
    term1 = -(y + 47) * np.sin(np.sqrt(abs(x/2 + (y + 47))))
    term2 = -x * np.sin(np.sqrt(abs(x - (y + 47))))
    return term1 + term2

# Domain boundaries
x_min, x_max = -512, 512
y_min, y_max = -512, 512

# Generating a meshgrid for plotting
x_values = np.linspace(x_min, x_max, 1000)
y_values = np.linspace(y_min, y_max, 1000)
X, Y = np.meshgrid(x_values, y_values)
Z = eggholder_function(X, Y)

# Monte Carlo sampling
num_samples = 50
x_samples = np.random.uniform(x_min, x_max, num_samples)
y_samples = np.random.uniform(y_min, y_max, num_samples)
mc_function_values = eggholder_function(x_samples, y_samples)

# Plotting
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

# Surface plot of the Eggholder function
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.5)

# Scatter plot of the Monte Carlo samples
ax.scatter(x_samples, y_samples, mc_function_values, color='red', s=50)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Eggholder Function')

plt.title("Monte Carlo Sampling of the Eggholder Function")
plt.show()
```

## Adaptive Sampling

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.stats import qmc

# Define the Eggholder function
def eggholder_function(x, y):
    term1 = -(y + 47) * np.sin(np.sqrt(abs(x/2 + (y + 47))))
    term2 = -x * np.sin(np.sqrt(abs(x - (y + 47))))
    return term1 + term2

# Domain boundaries
x_min, x_max = -512, 512
y_min, y_max = -512, 512

# Generate initial random samples
def initial_samples(num_samples=20):
    samples = np.random.rand(num_samples, 2)
    samples_scaled = qmc.scale(samples, (x_min, y_min), (x_max, y_max))
    return samples_scaled

# Evaluate the Eggholder function at the given samples
def evaluate_samples(samples):
    return eggholder_function(samples[:, 0], samples[:, 1])

# Adaptive sampling around the points of interest
def adaptive_sampling(initial_samples, iterations=5, samples_per_iter=10):
    current_samples = initial_samples
    for _ in range(iterations):
        function_values = evaluate_samples(current_samples)
        # Indices of points with highest absolute values (proxy for interest)
        top_indices = np.argsort(np.abs(function_values))[-10:]
        top_samples = current_samples[top_indices]

        # Generate new samples around these points
        new_samples = np.array([np.random.uniform(low=sample-50, high=sample+50, size=(samples_per_iter, 2)) for sample in top_samples])
        new_samples = new_samples.reshape(-1, 2)
        new_samples = np.clip(new_samples, [x_min, y_min], [x_max, y_max])

        # Add new samples to the current set
        current_samples = np.vstack((current_samples, new_samples))

    return current_samples

# Main function to run the adaptive sampling
def main():
    initial_samples_scaled = initial_samples(num_samples=20)
    adapted_samples = adaptive_sampling(initial_samples_scaled, iterations=5, samples_per_iter=5)

    # Evaluate function values for all samples
    all_function_values = evaluate_samples(adapted_samples)

    # Generating a meshgrid for plotting the Eggholder function
    x_values = np.linspace(x_min, x_max, 1000)
    y_values = np.linspace(y_min, y_max, 1000)
    X, Y = np.meshgrid(x_values, y_values)
    Z = eggholder_function(X, Y)

    # Plotting
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.5)
    ax.scatter(adapted_samples[:, 0], adapted_samples[:, 1], all_function_values, color='red', s=10)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Eggholder Function')
    plt.title("Adaptive Sampling of the Eggholder Function")
    plt.show()

if __name__ == "__main__":
    main()
```
